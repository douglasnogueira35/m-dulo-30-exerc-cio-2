```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# -----------------------------
# 1. Carregar dados
# -----------------------------
df = pd.read_csv("online_shoppers_intention.csv")

# -----------------------------
# 2. Seleção de variáveis para agrupamento
# -----------------------------
variaveis = [
    "Administrative", "Administrative_Duration",
    "Informational", "Informational_Duration",
    "ProductRelated", "ProductRelated_Duration",
    "PageValues", "SpecialDay", "Month", "Weekend"
]

X = df[variaveis].copy()

# -----------------------------
# 3. Tratamento de variáveis qualitativas
# -----------------------------
# Codificar Month (texto -> número)
le = LabelEncoder()
X["Month"] = le.fit_transform(X["Month"])
# Apply the same encoding to the original df for 'Month' before grouping
df["Month"] = le.fit_transform(df["Month"])

# Converter Weekend para 0/1
X["Weekend"] = X["Weekend"].astype(int)
# Apply the same conversion to the original df for 'Weekend' before grouping
df["Weekend"] = df["Weekend"].astype(int)

# -----------------------------
# 4. Padronização das variáveis
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -----------------------------
# 5. Construção do agrupamento hierárquico
# -----------------------------
Z = linkage(X_scaled, method="ward")

# Dendrograma (usar amostra para não travar)
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode="level", p=5) # Removido o fatiamento Z[:500]
plt.title("Dendrograma (amostra)")
plt.xlabel("Observações")
plt.ylabel("Distância")
plt.show()

# -----------------------------
# 6. Definição de grupos (3 e 4 clusters)
# -----------------------------
clusters_3 = fcluster(Z, 3, criterion="maxclust")
df["Cluster_3"] = clusters_3

clusters_4 = fcluster(Z, 4, criterion="maxclust")
df["Cluster_4"] = clusters_4

# -----------------------------
# 7. Análise descritiva dos grupos
# -----------------------------
print("Médias por cluster (3 grupos):")
print(df.groupby("Cluster_3")[variaveis].mean())

print("\nMédias por cluster (4 grupos):")
print(df.groupby("Cluster_4")[variaveis].mean())

# -----------------------------
# 8. Avaliação dos resultados
# -----------------------------
print("\nAvaliação com BounceRates e Revenue (3 grupos):")
print(df.groupby("Cluster_3")[["BounceRates", "Revenue"]].mean())

print("\nAvaliação com BounceRates e Revenue (4 grupos):")
print(df.groupby("Cluster_4")[["BounceRates", "Revenue"]].mean())
```

Interpretação dos 4 grupos
- Cluster 1 — Compradores decididos
- Mesmo perfil do cluster 1 anterior.
- Revenue: ~29%.
👉 Nome: “Compradores decididos”.
- Cluster 2 — Curiosos informativos
- Poucos acessos, mas maior proximidade a datas especiais (SpecialDay alto).
- Revenue: ~5,7%.
👉 Nome: “Visitantes de datas especiais”.
- Cluster 3 — Fim de semana exploradores
- Navegação moderada, mas 100% em fim de semana.
- Revenue: ~15%.
👉 Nome: “Exploradores de fim de semana”.
- Cluster 4 — Exploradores regulares
- Navegação intermediária, valores de página mais altos.
- Revenue: ~15,5%.
👉 Nome: “Exploradores regulares”.

🎯 Conclusão
- O grupo mais propenso à compra é claramente o Cluster 1 — Compradores decididos, com quase 30% de conversão.
- Os grupos menos propensos são os Curiosos/Visitantes ocasionais, com apenas ~5%.
- Os demais grupos ficam próximos da média (~15%).




```python
import matplotlib.pyplot as plt

# Gráfico comparando taxa de compra por cluster
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Clusters com 3 grupos
df.groupby("Cluster_3")["Revenue"].mean().plot(kind="bar", ax=ax[0], color="green")
ax[0].set_title("Taxa de compra por cluster (3 grupos)")
ax[0].set_ylabel("Taxa de compra média")

# Clusters com 4 grupos
df.groupby("Cluster_4")["Revenue"].mean().plot(kind="bar", ax=ax[1], color="blue")
ax[1].set_title("Taxa de compra por cluster (4 grupos)")
ax[1].set_ylabel("Taxa de compra média")

plt.tight_layout()
plt.show()
```
